```python
import pandas as pd
import os
```

```python
caminho = r"F:\Ponto Data\Reportagens\Cinema Ponta Grossa\Bilheteria di√°ria"
```

```python
arquivos = [f for f in os.listdir(caminho) if f.endswith('.csv')]
print("Arquivos encontrados:", arquivos)
```

```python
for arquivo in arquivos:
    caminho_arquivo = os.path.join(caminho, arquivo)
    print(f"Lendo: {arquivo}")
```

```python
df = pd.read_csv(caminho_arquivo, sep=None, engine='python')
```

```python
# Verifica se a coluna existe
'MUNICIPIO_SALA_COMPLEXO' in df.columns
```

```python
df_filtrado = df[df['MUNICIPIO_SALA_COMPLEXO'].str.contains('Ponta Grossa', case=False, na=False)]
```

```python
df_filtrado.head()
```

```python
lista_df = []

for arquivo in arquivos:
    caminho_arquivo = os.path.join(caminho, arquivo)
    print("Lendo:", arquivo)
    
    df = pd.read_csv(caminho_arquivo, sep=None, engine='python')
    
    # Filtra ‚ÄúPonta Grossa‚Äù
    df_filtrado = df[df['MUNICIPIO_SALA_COMPLEXO'].str.contains('Ponta Grossa', case=False, na=False)]
    
    lista_df.append(df_filtrado)

# Junta tudo
df_final = pd.concat(lista_df, ignore_index=True)
print("Total de linhas filtradas:", len(df_final))
```

```python
saida = os.path.join(caminho, "bilheteria_ponta_grossa.csv")
df_final.to_csv(saida, index=False, encoding='utf-8-sig')

print("‚úÖ Arquivo salvo em:", saida)
```
